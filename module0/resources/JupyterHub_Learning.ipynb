**JupyterHub**, as previously mentioned, is the centralized web application where students complete learning activities.

## Overview and Motivation

As a student in MU’s Data Science and Analytics (DSA) program, you’ll work with a wide array of computing resources, tools, and techniques, including:

  * The Python programming language, along with its libraries and ecosystem
  * R, with its own statistical libraries and extensions
  * Various types of databases, including relational (SQL), NoSQL, graph, and spatiotemporal databases
  * Cluster and cloud computing platforms

The MU JupyterHub environment serves as the main gateway for accessing all these computing resources. Through this environment, you will have access to Python, R, and server terminal interfaces, enabling you to:

  * Utilize programming languages and their libraries
  * Connect to and query different types of databases
  * Orchestrate data processing tasks via APIs and command-line tools on cloud or cluster computing infrastructures

Additional resources such as AWS, Google Cloud, and the MU NSF Supercomputer host large datasets and provide powerful compute capabilities. These cloud and supercomputing resources will be accessed through the JupyterHub environment.

For example, some learning activities will use Python notebooks within JupyterHub to interact with Google Cloud resources through the GCloud Python API. In other cases, you’ll use R notebooks that leverage the Google BigQuery API, allowing the Google Cloud to serve as a query interface and data repository. In all scenarios, you’ll remain within the MU JupyterHub environment, with API connections extending to external resources as needed (such as using the BOTO3 library for Python-to-AWS connectivity). 

Certain activities will also guide you in preparing and submitting data processing jobs to supercomputing clusters using Python scripts and terminal commands.

Our program features a close integration and workflow management system between JupyterHub and Gitlab, streamlining course modules and activities for the DSA curriculum.

# Gitlab

Gitlab is a centralized deployment of Git Version Control System repositories.

# Git

Version control systems (VCS) are utilized as change management systems for a variety of electronic items, such as source code, configuration files, data files, documentation, etc. Git is a very powerful VCS that is rapidly expanding in use through the technology realms. Read more about Git [here](https://git-scm.com/about/). Instead of having lessons on Git and version control, you will learn it by continual use throughout the program.

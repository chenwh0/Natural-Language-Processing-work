# Saving your work using Git

This notebook is a light walk-through of the steps to push (save) your work back to your repository (repo for short) in the Git version control server.

## Terminal Access to the Server
To perform these steps, you have to use a Linux terminal session.
JupyterHub provides a basic terminal for us.

##  Review our work    

Switch to the course folder and and look at the repo status: 
```
cd sp25dsa8010*
git status
```

Any files you have changed will show up in red: **Look for files in red, labeled as "modified"**.

![JH Terminal](../images/JH_modified_status.png)


##  Stage Changes
```
git add PATH_TO_FILE
```
  * `PATH_TO_FILE` should be the actual file you want to commit!
  
NOTE: You need to add only the _.ipynb_ files.

It is a good practice to add everything in your course folder so that you have a backup in your repo. 

```
git add . --all
```

    
##  Create a commit (i.e., a save point)
```
git commit -m "This is my Module 1 work"
```
    
    
## Send these changes to the server for safe keeping (i.e., publish)
```
git push origin master 
```
   
## Confirm our working folder is clean and all work is tracked
```
git status
```
